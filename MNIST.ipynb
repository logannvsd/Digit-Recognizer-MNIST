```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Conv2D,MaxPooling2D,Flatten,PReLU
from tensorflow.keras.preprocessing.image import ImageDataGenerator

import numpy as np
from matplotlib import image
import pickle
```

```python
model = Sequential()
model.add(Conv2D(128, (3,3), padding='same', input_shape = (28, 28, 4), kernel_initializer="he_normal"))
model.add(PReLU())
model.add(MaxPooling2D((2, 2)))

model.add(Conv2D(128, (3,3), padding='same', kernel_initializer="he_normal"))
model.add(PReLU())
model.add(MaxPooling2D((2, 2)))

model.add(Conv2D(128, (3,3), padding='same', kernel_initializer="he_normal"))
model.add(PReLU())
model.add(MaxPooling2D((2, 2)))

model.add(Flatten())
model.add(Dense(10,activation = 'softmax'))
model.summary()
#number of layers, number of kernels, epochs
```
